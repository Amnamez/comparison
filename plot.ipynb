```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- 1. Data from your table (Model names and their scores) ---
data = {
    'Model': ['TF-IDF + SVM', 'Naive Bayes', 'Random Forest', 'MY-BERT'],
    'Precision': [0.08, 0.05, 0.40, 0.86],
    'Recall':    [0.08, 0.05, 0.12, 0.82],
    'F1 Score':  [0.077, 0.051, 0.008, 0.85],
    'Accuracy':  [0.9998, 0.9988, 0.9998, 0.95]
}

df = pd.DataFrame(data)

# --- 2. First Plot: Precision, Recall, F1 (Smaller size) ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(8, 5))  # Changed from (10,6) to (8,5)

x = np.arange(len(df['Model']))
width = 0.25

rects1 = ax.bar(x - width, df['Precision'], width, label='Precision', color='#1f77b4')
rects2 = ax.bar(x, df['Recall'], width, label='Recall', color='#ff7f0e')
rects3 = ax.bar(x + width, df['F1 Score'], width, label='F1 Score', color='#2ca02c')

ax.set_xlabel('Model', fontsize=10, fontweight='semibold')
ax.set_ylabel('Score', fontsize=10, fontweight='semibold')
ax.set_title('Comparison of Models on Keyword Extraction', fontsize=11, fontweight='bold', pad=15)
ax.set_xticks(x)
ax.set_xticklabels(df['Model'], rotation=15, ha='right', fontsize=8)
ax.set_ylim(0, 1.0)
ax.legend(loc='upper left', frameon=True, fancybox=True, shadow=True, fontsize=8)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 2),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=6)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

ax.axvspan(x[3] - 0.5, x[3] + 0.5, alpha=0.1, color='green')

plt.tight_layout()
plt.show()

# --- 3. Second Plot: Accuracy (Even smaller) ---
fig2, ax2 = plt.subplots(figsize=(7, 4))  # Changed from (10,5) to (7,4)

ax2.bar(df['Model'], df['Accuracy'], color='#9467bd')
ax2.set_ylabel('Accuracy', fontsize=10, fontweight='semibold')
ax2.set_title('Comparison of Model Accuracy', fontsize=11, fontweight='bold')
ax2.set_ylim(0.94, 1.001)
ax2.tick_params(axis='x', rotation=15, labelsize=8)
ax2.tick_params(axis='y', labelsize=8)

for i, (model, acc) in enumerate(zip(df['Model'], df['Accuracy'])):
    ax2.text(i, acc + 0.0005, f'{acc:.4f}', ha='center', va='bottom', fontsize=7)

plt.tight_layout()
plt.show()
```
